```python
import re
import pandas as pd

# --- remove "**N. Risk Drivers**" + its bullets + the following underline line only ---
RISK_DRIVERS_BLOCK = re.compile(
    r'''(?imx)                # ignorecase, multiline, verbose
    ^\*\*\s*(?:\d+\.\s*)?     # ** and optional "N."
    risk\s*drivers\s*\*\*     # "Risk Drivers"**
    (?:\s*\r?\n[ \t]*)*       # optional blank line(s)
    (?:                       # one or more bullet lines
       [ \t]*(?:[•\-\*]\s.*)\r?\n
    )+
    [ \t]*_+\s*\r?\n?         # the immediate underline divider line
    ''',
)

# rename "**5. Compared Transactions**" -> "**4. Compared Transactions**"
RENAME_CT = re.compile(r'\*\*\s*5\.\s*Compared\s*Transactions\s*\*\*', flags=re.IGNORECASE)

def strip_risk_drivers_and_renumber(text: str) -> str:
    if not isinstance(text, str):
        return text
    out = RISK_DRIVERS_BLOCK.sub('', text)
    out = RENAME_CT.sub('**4. Compared Transactions**', out)
    # tidy excessive blank lines introduced by removal
    out = re.sub(r'\n{3,}', '\n\n', out).strip()
    return out




df=pd.read_excel('data_before_db_update.xlsx')
df['llm_refined_explanation_without_risk_driver'] = df['llm_refined_explanation'].astype(str).apply(strip_risk_drivers_and_renumber)
#df_fixed.to_excel('data_for_db_update_without_risk_driver_17_10.xlsx',index=False)
```

```python
import re
import pandas as pd

def ensure_impact_areas_for_invoice_exceeds_po(
    df: pd.DataFrame,
    text_col: str = "llm_explanation",
    out_text_col: str = "updated_llm_explanation",
    subrisk_cols = ("sub_risk_1", "sub_risk_2", "sub_risk_3"),
    keywords = ("invoice exceeds po", "invoice amount or quantity exceeds po"),
    impacts_to_ensure = ("Profitability", "Cash Flow"),
):
    """
    If any of sub_risk_1/2/3 indicates 'Invoice Exceeds PO' (incl. the longer wording),
    make sure the '• **Impact Areas:** ...' bullet in llm_explanation includes
    'Profitability, Cash Flow'. Adds the missing ones, without duplicating existing items.

    - Preserves formatting:
        • **Impact Areas:** <comma-separated items>.
    - If the Impact Areas bullet is empty or '.', it will be replaced with the required items.
    - If the Impact Areas bullet is missing entirely, it appends a new bullet at the end.
    """

    df = df.copy()

    # Normalize check for sub-risk trigger
    def _has_invoice_exceeds(row) -> bool:
        for c in subrisk_cols:
            val = str(row.get(c, "") or "").strip().lower()
            if any(k in val for k in keywords):
                return True
        return False

    # Regex to find the Impact Areas bullet line (multiline)
    # Matches lines like: "• **Impact Areas:** something"
    impact_line_pat = re.compile(r"(^\s*•\s*\*\*Impact Areas:\*\*\s*)(.*)$", re.MULTILINE)

    def _normalize_items(s: str) -> list:
        # strip trailing period and whitespace
        s = s.strip()
        if s.endswith("."):
            s = s[:-1]
        # split by comma, strip parts, drop empties
        parts = [p.strip() for p in s.split(",")]
        return [p for p in parts if p]

    def _recompose(items: list) -> str:
        # Join with ", " and add trailing period.
        return (", ".join(items) + ".") if items else "."

    def _update_impact_line(existing_tail: str) -> str:
        """
        existing_tail is the text after '• **Impact Areas:** ' on that line.
        Return the new tail with ensured impacts.
        """
        existing_items = _normalize_items(existing_tail)
        # Make a case-insensitive set to check presence, but preserve original case of existing
        lower_existing = {x.lower() for x in existing_items}

        new_items = list(existing_items)  # preserve order of existing
        for imp in impacts_to_ensure:
            if imp.lower() not in lower_existing:
                new_items.append(imp)

        # If nothing existed (like '.' or empty), ensure the required two are present
        if not new_items:
            new_items = list(impacts_to_ensure)

        return _recompose(new_items)

    def _ensure_impact_areas(text: str) -> str:
        # Try to find and update the existing Impact Areas line
        m = impact_line_pat.search(text)
        if m:
            prefix, tail = m.group(1), m.group(2)
            new_tail = _update_impact_line(tail)
            # Replace only this one line
            return text[:m.start()] + (prefix + new_tail) + text[m.end():]

        # If there's no Impact Areas line at all, append a new one at the end
        add_line = f"\n• **Impact Areas:** {', '.join(impacts_to_ensure)}."
        return text.rstrip() + add_line

    out = []
    for _, row in df.iterrows():
        text = row.get(text_col, "")
        if _has_invoice_exceeds(row) and isinstance(text, str) and text.strip():
            out.append(_ensure_impact_areas(text))
        else:
            out.append(text)

    df[out_text_col] = out
    return df

# -------------------------
# Example usage:
# Replace NaN / NA with None in sub-risk columns
for c in ["sub_risk_1", "sub_risk_2", "sub_risk_3"]:
    if c in new_df_final.columns:
        new_df_final[c] = new_df_final[c].apply(lambda x: None if pd.isna(x) else x)

df_updated = ensure_impact_areas_for_invoice_exceeds_po(new_df,
        text_col="llm_refined_explanation_without_risk_driver",
         out_text_col="llm_refined_explanation_without_risk_driver_updated")
```

```python
import pandas as pd
import numpy as np
import math

# === Your desired target schema (as you gave) ================================
TARGET_SCHEMA = {
    "purch_doc_no_src_po": "object",
    "purch_doc_item_no_src_po": "object",
    "pr_no_src_po": "object",
    "pr_item_no_src_po": "object",
    "requester_name_src_po": "object",
    "doc_change_date_src_po": "datetime64[ns]",
    "company_code_src_po": "object",
    "p2o_unit_conv_denom_src_po": "float64",
    "o2b_unit_conv_denom_src_po": "float64",
    "p2o_unit_conv_num_src_po": "float64",
    "o2b_unit_conv_num_src_po": "float64",
    "po_item_del_flag_src_po": "object",
    "gr_indicator_src_po": "object",
    "gr_invoice_verif_flag_src_po": "object",
    "gross_val_po_curr_src_po": "float64",
    "inv_receipt_indicator_src_po": "object",
    "material_type_src_po": "object",
    "material_no_src_po": "object",
    "matl_group_src_po": "object",
    "net_price_doc_curr_src_po": "float64",
    "net_val_po_curr_src_po": "float64",
    "order_uom_src_po": "object",
    "quantity_src_po": "float64",
    "plant_src_po": "object",
    "short_text_src_po": "object",
    "target_qty_src_po": "float64",
    "outline_agrmt_tgt_val_doc_curr_src_po": "float64",
    "reqmt_tracking_no_src_po": "object",
    "principal_purch_agrmt_item_no_src_po": "float64",
    "base_id_src_po": "object",
    "purch_doc_no_hpd_po": "object",
    "vendor_or_creditor_acct_no_hpd_po": "object",
    "purch_doc_category_hpd_po": "object",
    "company_code_hpd_po": "object",
    "currency_hpd_po": "object",
    "object_created_by_hpd_po": "object",
    "doc_change_date_hpd_po": "datetime64[ns]",
    "control_indicator_hpd_po": "object",
    "purch_doc_date_hpd_po": "datetime64[ns]",
    "principal_purch_agrmt_no_hpd_po": "object",
    "exchange_rate_hpd_po": "float64",
    "purch_group_hpd_po": "object",
    "purch_org_hpd_po": "object",
    "pymnt_terms_hpd_po": "object",
    "processing_status_hpd_po": "object",
    "doc_release_incompl_flag_hpd_po": "object",
    "release_indicator_hpd_po": "object",
    "release_status_hpd_po": "object",
    "rfq_status_hpd_po": "object",
    "resp_vendor_salesperson_hpd_po": "object",
    "on_release_total_value_hpd_po": "float64",
    "purch_doc_type_hpd_po": "object",
    "vendor_or_creditor_acct_no": "object",
}

# === Helper casters ==========================================================
def _to_object_clean(series: pd.Series) -> pd.Series:
    """
    Convert any series to plain Python-object strings.
    - Missing -> None
    - Floats like 123.0 -> "123"
    - Leaves non-missing strings as-is
    """
    def _clean_one(v):
        if pd.isna(v):
            return None
        # ints/floats -> avoid "4000003557.0"
        if isinstance(v, (int, np.integer)):
            return str(int(v))
        if isinstance(v, (float, np.floating)):
            if math.isfinite(float(v)) and float(v).is_integer():
                return str(int(v))
            return str(v)
        # anything else
        return str(v)

    # ensure we're writing into a new object dtype container
    out = series.map(_clean_one)
    return out.astype("object")

def _to_float64(series: pd.Series) -> pd.Series:
    return pd.to_numeric(series, errors="coerce").astype("float64")

def _to_datetime(series: pd.Series) -> pd.Series:
    return pd.to_datetime(series, errors="coerce")

# === Main function: apply the schema IN PLACE ================================
def apply_schema_inplace(df: pd.DataFrame, schema: dict, verbose: bool = True):
    """
    For each (col -> dtype) in schema, cast IN PLACE.
    - object -> safe stringify (with None for missing, strip .0)
    - float64 -> numeric (coerce)
    - datetime64[ns] -> to_datetime (coerce)
    Other dtypes are ignored here (extend if needed).
    Columns not in df are skipped.
    """
    for col, target in schema.items():
        if col not in df.columns:
            if verbose: print(f"[skip] '{col}' not in DataFrame")
            continue
        before = str(df[col].dtype)
        try:
            if target == "object":
                df.loc[:, col] = _to_object_clean(df[col])
            elif target == "float64":
                df.loc[:, col] = _to_float64(df[col])
            elif target == "datetime64[ns]":
                df.loc[:, col] = _to_datetime(df[col])
            else:
                # If you later add int64 or others, handle here as needed.
                df.loc[:, col] = df[col].astype(target)
            after = str(df[col].dtype)
            if verbose:
                print(f"[ok] {col}: {before} -> {after}")
        except Exception as e:
            print(f"[fail] {col}: {before} -> {target} | {e}")

# ================== How to run ==============================================
# Example:
apply_schema_inplace(new_df, TARGET_SCHEMA, verbose=True)
# Now df.info() should show the exact target dtypes for those columns.
```
